# 리뷰수, 평점 업데이트

```python
#이전 데이터 불러오기
preData = pd.read_csv('all_location.csv', header=None)
#칼럼 추가
preData.columns = ['매장','테마','평점','리뷰수', '블로그 리뷰','난이도', '지역','id']
#새 데이터 불러오기
newData = pd.read_csv('all_location_최종.csv')

#업데이트 루프
for row in preData.iterrows():
  b = newData.loc[(newData['매장']==row[1]['매장']) & (newData['테마']==row[1]['테마'])]
  if len(b)==1:
    score = b['평점'].values
    review = b['리뷰수'].values
    preData.loc[(preData['매장']==row[1]['매장']) & (preData['테마']==row[1]['테마']), '평점'] = score[0]
    preData.loc[(preData['매장']==row[1]['매장']) & (preData['테마']==row[1]['테마']),'리뷰수'] = review[0]
```

```python
preData = preData.sort_values(by='평점', ascending=False, ignore_index=True)
preData.to_csv('all_location_update.csv', index=False)
```
